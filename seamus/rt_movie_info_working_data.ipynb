We want to separate out the multiple genres listsed under the "genre" column.

```python
movie_info_df['genre'].value_counts()
```

```python
# Split the genre column into separate strings based on the "|" separator
split_genres = movie_info_df['genre'].str.split('|', expand=True)
```

```python
# Rename the columns to indicate the genre
split_genres.columns = [f"genre_{i+1}" for i in range(split_genres.shape[1])]
```

```python
# Combine the original dataframe with the split genres dataframe
movie_info_df = pd.concat([movie_info_df, split_genres], axis=1)
```

```python
movie_info_df['genre'] = movie_info_df['genre'].astype(str)
```

```python
movie_info_df['genre_list'] = movie_info_df['genre'].apply(lambda x: x.split('|'))
```

```python
movie_info_df['genre_list'][0]
```

```python
movie_info_df = movie_info_df.explode('genre_list')
```

```python
movie_info_df = movie_info_df.drop(['genre_1', 'genre_2', 'genre_3', 'genre_4', 'genre_5', 'genre_6', 'genre_7'], axis=1)
```
