```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bom_df = pd.read_csv('/Users/seamuswalsh/Flatiron/Course Materials/Phase 1/Phase_1_Project/Exploratory-Data-Analysis-Using-Movie-Data/seamus/Movie Project Data/bom.movie_gross.csv')
```

```python
bom_df['domestic_gross'] = bom_df['domestic_gross'].fillna(bom_df['domestic_gross'].median())
```

```python
bom_df = bom_df.drop('foreign_gross', axis=1)
```

```python
bom_df.dropna(inplace = True)
```

```python
bom_df
```

```python
#bom_df['studio'] = bom_df['studio'].str.replace('$', '')
```

```python
sns.scatterplot(x = 'studio', y = 'domestic_gross', data = bom_df )
plt.show()
```

```python
tn_df = pd.read_csv('/Users/seamuswalsh/Flatiron/Course Materials/Phase 1/Phase_1_Project/Exploratory-Data-Analysis-Using-Movie-Data/seamus/Movie Project Data/tn.movie_budgets.csv')
tn_df
```

```python
df3 = pd.merge(tn_df, bom_df, left_on='movie', right_on='title')
df3
```

```python
df3['production_budget'] = df3['production_budget'].str.replace('$', '')
df3['production_budget'] = df3['production_budget'].str.replace(',', '')
df3['domestic_gross_x'] = df3['domestic_gross_x'].str.replace('$', '')
df3['domestic_gross_x'] = df3['domestic_gross_x'].str.replace(',', '')
df3['worldwide_gross'] = df3['worldwide_gross'].str.replace('$', '')
df3['worldwide_gross'] = df3['worldwide_gross'].str.replace(',', '')
```

```python
df3 = df3.astype({'production_budget':'int'})
df3 = df3.astype({'domestic_gross_x':'int'})
df3 = df3.astype({'worldwide_gross':'int'})
```

```python
df3['release_date'] = pd.to_datetime(df3['release_date'], infer_datetime_format=True) 
```

```python
df3['return_on_investment'] = (df3['worldwide_gross']/df3['production_budget'])*100
```

```python
# Removing ROI values that are $0.
df3 = df3[df3['return_on_investment'] != 0]
```

```python
df3
```

```python
df3['release_month'] = pd.DatetimeIndex(df3['release_date']).month
df3
```

```python
df_sum = df3.groupby('release_month').sum('domestic_gross_y')
df_sum
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

x = df3['studio']
y = df3['return_on_investment']

ax.scatter(x,y)
ax.set_title('Relationship Between studio and Return on Investment (ROI)')
ax.set_xlabel('studio')
ax.set_ylabel('Return on Investment (ROI)');
```

```python
df3['best_roi'] = np.where(df3['return_on_investment'] > 500, 'good', 'bad')
```

```python
df3['best_roi'].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

x = df3['best_roi']
y = df3['production_budget']

ax.scatter(x,y)
ax.set_title('Relationship Between studio and Production Budget')
ax.set_xlabel('studio')
ax.set_ylabel('Production Budget');
```

```python
sns.boxplot(x = 'best_roi', y = 'production_budget', 
            data = df3, orient = 'v')
plt.show()
```

```python
df3['return_on_investment'].describe()
```

```python
q1 = df3['return_on_investment'].quantile(0.25)
q3 = df3['return_on_investment'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Create a new column that contains only the non-outlier values
df3['best_roi_studios'] = df3.loc[(df3['return_on_investment'] >= lower_bound) & (df3['return_on_investment'] <= upper_bound), 'return_on_investment']
df3
```

```python
sns.boxplot(x = 'best_roi', y = 'best_roi_studios', 
            data = df3, orient = 'v')
plt.show()
```

```python
df3['studio'].value_counts()


# Define the threshold for counting as a boolean value
threshold = 100

# Count the occurrences of each value in column A
value_counts = df3['studio'].value_counts()

# Create a new column that assigns True/False based on the count
df3['top_studios'] = df3['studio'].apply(lambda x: value_counts[x] >= threshold)

# Convert the True/False values to 1/0 integers if desired
df3['top_studios'] = df3['top_studios'].astype(int)

df3['top_studios'].value_counts()
```

```python
sns.scatterplot(x = 'return_on_investment', element = 'step', data = df3)
plt.show()
```
