```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movie_budgets_df = pd.read_csv('tn.movie_budgets.csv')
movie_budgets_df

bom_df = pd.read_csv('bom.movie_gross.csv')
bom_df
```

```python
movie_budgets_df['release_date'] =  pd.to_datetime(movie_budgets_df['release_date'], infer_datetime_format=True)
movie_budgets_df
```

```python
df3 = pd.merge(movie_budgets_df, bom_df, left_on='movie', right_on='title')
df3
```

```python
df3['release_date'] =  pd.to_datetime(df3['release_date'], infer_datetime_format=True)
```

```python
df3.sort_values(by = ['release_date'], ascending = False).head(50)
```

```python
df3['release_month'] = pd.DatetimeIndex(df3['release_date']).month
df3
```

```python
df_sum = df3.groupby('release_month').sum('domestic_gross_y')
df_sum['domestic_gross_y']
```

```python
df_sum['domestic_gross_y'].plot(kind = 'bar')
plt.show()
```
